### **4. Visualizing Car Fuel Type Feature.**

```python
def categorical_visualization(cols):
    plt.figure(figsize=(20,8))
    plt.subplot(1,3,1)
    sns.countplot(x=cols,data=df,palette="Set2",order=df[cols].value_counts().index)
    plt.title(f"{cols} Distribution",pad=10,fontweight="black",fontsize=18)
    plt.xticks(rotation=90)

    plt.subplot(1,3,2)
    sns.boxplot(x=cols,y="price",data=df,palette="Set2")
    plt.title(f"{cols} vs Price",pad=20,fontweight="black",fontsize=18)
    plt.xticks(rotation=90)
    
    plt.subplot(1,3,3)
    x=pd.DataFrame(df.groupby(cols)["price"].mean().sort_values(ascending=False))
    sns.barplot(x=x.index,y="price",data=x,palette="Set2")
    plt.title(f"{cols} vs Average Price",pad=20,fontweight="black",fontsize=18)
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()

categorical_visualization("fueltype")
```

Analysis:
1. เราเห็นได้อย่างชัดเจนว่ารถยนต์ที่มีระบบเชื้อเพลิงแก๊สขายเป็นส่วนใหญ่จากกราฟที่ 2 ทำให้เข้าใจได้ว่ารถยนต์ระบบ Gas มีจำหน่ายในทุกช่วงราคา
2. รถยนต์เชื้อเพลิงทั้งสองประเภทมีราคารถยนต์ส่วนใหญ่ใกล้เคียงกับราคาเฉลี่ยของรถยนต์ แต่รถยนต์ประเภทเชื้อเพลิงแก๊สก็มีรถยนต์ที่มีราคาสูง
3. จากกราฟที่ 3 เราพอจะสรุปได้ว่าราคาเฉลี่ยของรถยนต์ประเภท Gas มีราคาต่ำกว่ารถยนต์ประเภท Diesel จึงบอกได้เลยว่าลูกค้าชอบรถที่กินน้ำมันราคาถูกกว่า

# **Data Preprocessing**

```python
z = round(df.groupby(["CompanyName"])["price"].agg(["mean"]),2).T
z
```

```python
df = df.merge(z.T,how="left",on="CompanyName")
df
```

```python
bins = [0,10000,20000,40000]
cars_bin=['Budget','Medium','Highend']
df['CarsRange'] = pd.cut(df['mean'],bins,right=False,labels=cars_bin)
df.head()
```

### **2. Encode Data เก็บเป็น Dict**

```python
def label_encode(data):
    unique_labels = list(set(data))
    label_dict = {label: index for index, label in enumerate(unique_labels)}
    encoded_data = [label_dict[item] for item in data]
    return encoded_data, label_dict
```

```python
new_df['fueltype'], fueltype_level_dict = label_encode(new_df['fueltype'])
new_df['aspiration'], aspiration_level_dict = label_encode(new_df['aspiration'])
new_df['doornumber'], doornumber_level_dict = label_encode(new_df['doornumber'])
new_df['carbody'], carbody_level_dict = label_encode(new_df['carbody'])
new_df['drivewheel'], drivewheel_level_dict = label_encode(new_df['drivewheel'])
new_df['enginetype'], enginetype_level_dict = label_encode(new_df['enginetype'])
new_df['cylindernumber'], cylindernumber_level_dict = label_encode(new_df['cylindernumber'])
new_df['fuelsystem'], fuelsystem_level_dict = label_encode(new_df['fuelsystem'])
new_df['CarsRange'], CarsRange_level_dict = label_encode(new_df['CarsRange'])
new_df
```

```python
training_score = []
testing_score = []
```

```python
def model_prediction(model):
    model.fit(x_train,y_train)
    x_train_pred = model.predict(x_train)
    x_test_pred = model.predict(x_test)
    
    y_train_np = np.array(y_train)
    y_test_np = np.array(y_test)
    
    a = r2_score(y_train,x_train_pred)*100
    b = r2_score(y_test,x_test_pred)*100
    training_score.append(a)
    testing_score.append(b)
    
    cosine_similarity_train = cosine_similarity(x_train_pred.reshape(1, -1), y_train_np.reshape(1, -1))[0][0]
    cosine_similarity_test = cosine_similarity(x_test_pred.reshape(1, -1), y_test_np.reshape(1, -1))[0][0]
    
    pearson_corr_train, _ = pearsonr(x_train_pred, y_train_np)
    pearson_corr_test, _ = pearsonr(x_test_pred, y_test_np)
    # Calculate MSE
    mse_train = mean_squared_error(y_train, x_train_pred)
    mse_test = mean_squared_error(y_test, x_test_pred)
    
    # Calculate MAE
    mae_train = mean_absolute_error(y_train, x_train_pred)
    mae_test = mean_absolute_error(y_test, x_test_pred)
    
    print(f"r2_Score of {model} model on Training Data is:",a)
    print(f"r2_Score of {model} model on Testing Data is:",b)
    print(f"Cosine Similarity of {model} model on Training Data is: {cosine_similarity_train}")
    print(f"Cosine Similarity of {model} model on Testing Data is: {cosine_similarity_test}")
    print(f"Pearson Correlation of {model} model on Training Data is: {pearson_corr_train}")
    print(f"Pearson Correlation of {model} model on Testing Data is: {pearson_corr_test}")
    print(f"MSE of {model} model on Training Data is: {mse_train}")
    print(f"MSE of {model} model on Testing Data is: {mse_test}")
    print(f"MAE of {model} model on Training Data is: {mae_train}")
    print(f"MAE of {model} model on Testing Data is: {mae_test}")
```

```python
lr = LinearRegression()
model_prediction(lr)
```
